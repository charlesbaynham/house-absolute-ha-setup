```python
import numpy as np
import matplotlib.pyplot as plt
```

Goal
---



Algorithm for a 2D plane which, given two points (vectors $\vec{x_1}$ and $\vec{x_2}$), determines if the line between them passes through a circle defined by centre vector $\vec{c_0}$ and radius $r$.

Idea:

If the line passes through the circle, it'll pass through the line that's perpendicular to the vector $\vec c - \vec {x_1}$. Therefore, find this line and then see if the two lines intersect.

Checking if two lines intersect
----

Given $\vec{x_1}, \vec{x_2}, \vec{y_1}, \vec{y_2}$, do the lines X and Y intersect?

Parameterise the lines in terms of $\alpha$ and $\beta$.

$$
\vec{X}(\alpha) = \vec{x_1} + \alpha (\vec{x_2} - \vec{x_1}) \\
\vec{Y}(\beta) = \vec{y_1} + \beta (\vec{y_2} - \vec{y_1})
$$

Solve $\vec{X} = \vec{Y}$ for $\alpha, \beta$.

$$
\alpha (\vec{x_2} - \vec{x_1}) - \beta (\vec{y_2} - \vec{y_1}) = \vec{y_1} - \vec{x_1}
$$

$$
\implies\quad
\begin{bmatrix}
(\vec{x_2} - \vec{x_1}) & (\vec{y_1} - \vec{y_2})
\end{bmatrix}
\begin{bmatrix}
\alpha \\ \beta
\end{bmatrix}
=
M
\begin{bmatrix}
\alpha \\ \beta
\end{bmatrix}
=
\vec{y_1} - \vec{x_1}
$$

$$
\begin{bmatrix}
\alpha \\ \beta
\end{bmatrix}
=
M^{-1}
(\vec{y_1} - \vec{x_1})
$$

if

$$
M
=
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

then

$$
M^{-1} = 
\frac{1}{ad-bc} \begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix}
$$

Example. Does the line from (-1, -1) to (1, 0) cross the line from (-1, 0) to (1, -1)? (yes it does)


```python
x1 = np.array([-1,-1])
x2 = np.array([1,0])
y1 = np.array([-1,0])
y2 = np.array([1,-1])

def lines_cross(x1,x2,y1,y2) -> bool:

    a = x2[0]-x1[0]
    b = y1[0]-y2[0]
    c = x2[1]-x1[1]
    d = y1[1]-y2[1]

    Minv = 1 / (a*d - b*c) * np.array([[d, -b],[-c,a]])
    alpha_beta = Minv @ (y1 - x1)
    
    plt.plot([x1[0], x2[0]], [x1[1], x2[1]])
    plt.plot([y1[0], y2[0]], [y1[1], y2[1]])

    return all((0 <= alpha_beta )& (alpha_beta <= 1))

lines_cross(x1,x2,y1,y2)
```

How about (-1, -1) to (1, 1) and (-1, 0) to (1, 1.1)? This one doesn't

```python
x1 = np.array([-1,-1])
x2 = np.array([1,1])
y1 = np.array([-1,0])
y2 = np.array([1,1.1])

lines_cross(x1,x2,y1,y2)
```

Great, that works. Now how do I get the line across the circle, C?

$$
C(\gamma) = \vec{c_1} + \gamma (\vec{c_2} - \vec{c_1})
$$

Get the unit vector that's perpendicular to $x_1 \rightarrow c$

$$
\hat u = 
\frac{1}{|\vec{x_1} - \vec{c}|}
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
(\vec{x_1} - \vec{c})
$$

The line is therefore

$$
\vec{c_1} = \vec{c} - r \hat u 
\\
\vec{c_2} = \vec{c} + r \hat u 
$$

```python
def get_circle_points(c,r,x1):
    v = x1-c
    u = (1 / np.sqrt(np.dot(v,v))) * np.array([[0,1],[-1,0]]) @ v

    c1 = c - r * u
    c2 = c + r * u
    
    return c1,c2
```

Try a line from (0,0) to (2,2) with a circle centred on (2,0) with radius 1 - not crossing.

```python
c = np.array([2,0])
r = 1
x1 = np.array([0,0])
x2 = np.array([2,2])

y1,y2 = get_circle_points(c,r,x1)
lines_cross(x1,x2,y1,y2)
```

Try a line from (0,0) to (2,2) with a circle centred on (2,0) with radius sqrt(2) - just crossing.

```python
c = np.array([2,0])
r = np.sqrt(2)
x1 = np.array([0,0])
x2 = np.array([2,2])

y1, y2 = get_circle_points(c,r,x1)
lines_cross(x1,x2,y1,y2)
```

Mistake - the line isn't the right one. The line shouldn't be perpendicular to $\vec c - \vec{x_1}$, it should be perpendicular to $\vec {x_2} - \vec{x_1}$.

$$
\implies\quad
\hat u = 
\frac{1}{|\vec{x_1} - \vec{x_2}|}
\begin{bmatrix}
0 & 1 \\
-1 & 0
\end{bmatrix}
(\vec{x_1} - \vec{x_2})
$$

```python
def get_circle_points(c,r,x1,x2):
    v = x1-x2
    u = (1 / np.sqrt(np.dot(v,v))) * np.array([[0,1],[-1,0]]) @ v

    c1 = c - r * u
    c2 = c + r * u
    
    return c1,c2
```

```python
c = np.array([2,0])
r = np.sqrt(2)
x1 = np.array([0,0])
x2 = np.array([2,2])

y1, y2 = get_circle_points(c,r,x1,x2)
lines_cross(x1,x2,y1,y2)
```

```python
c = np.array([2,0])
r = np.sqrt(2) - 0.1
x1 = np.array([0,0])
x2 = np.array([2,2])

y1, y2 = get_circle_points(c,r,x1,x2)
lines_cross(x1,x2,y1,y2)
```

```python
c = np.array([2,0])
r = np.sqrt(2) - 0.1
x1 = np.array([1,-2])
x2 = np.array([4,2])

y1, y2 = get_circle_points(c,r,x1,x2)
lines_cross(x1,x2,y1,y2)
```
